```python
import os
from IPython.display import HTML
from rdkit.Chem import PandasTools
import pandas as pd

def displaydf(df):
    return HTML(df.to_html(notebook=True))

ddir = 'data/'
df = pd.read_csv(os.path.join(ddir,'partial_data.csv'))
df.columns = ['smiles']
display(df)

PandasTools.AddMoleculeColumnToFrame(df,'smiles','mol',includeFingerprints=False)
# displaydf(df.head(2))
```

```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG = True

from graph_aug_utils import add_atom_to_mol
from rdkit_utils import show_atom_index, get_rwmol

def MolGraphFromSmiles(smiles):
    return get_rwmol( Chem.MolFromSmiles(smiles) )

df['length'] = df.smiles.apply(lambda x: MolGraphFromSmiles(x).GetNumAtoms())
df = df.sort_values('length',ascending=True)
# displaydf(df.head(5))

df_smalls = df[:20]
displaydf(df_smalls)
```

```python
from graph_aug_utils import randomly_add_atom
for _,row in df_smalls.iterrows():    
    print('- '*40,'\nORIGINAL MOLECULE')
    molgraph = MolGraphFromSmiles(row['smiles'])
    display(molgraph)
    print('AUGMENTED MOLECULES')
    for i in range(3):
        aidx, molgraph_new = randomly_add_atom(molgraph, 'S')
        display(aidx, molgraph_new)
        print(".  .  .  .  .")
```
