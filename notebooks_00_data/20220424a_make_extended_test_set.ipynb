```python
%load_ext autoreload
%autoreload 2
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
import warnings
warnings.filterwarnings("ignore")

import os
import pandas as pd
import sys
sys.path.insert(0, '/home/kat/Repos/SALSA/')

ddir = '/home/kat/Repos/SALSA/data/'
df = pd.read_csv(os.path.join(ddir,'chembl_valid_10to30atoms.csv'))
display(df)
```

```python
import random
random.seed(666)

N = 20000

idc = random.sample(range(len(_df)), N+N//10)
print(len(idc))
train_idc = random.sample(idc, N)
test_idc = [i for i in idc if i not in train_idc]

print(len(train_idc), len(test_idc))
print(set(train_idc).isdisjoint(set(test_idc)))
```

```python
# takes about 5-7 (?) mins to run!!

taken_idc = train_idc + test_idc
leftover_idc = [i for i in range(len(_df)) if i not in taken_idc]
```

```python
extended_idc = random.sample(leftover_idc, (2000*6)-2000)
bigtest_idc = extended_idc + test_idc
print(set(train_idc).isdisjoint(set(bigtest_idc)))
print(len(bigtest_idc))
```

```python
df_bigtest = df.iloc[bigtest_idc]
df_bigtest.to_csv(f'{ddir}model_ready/01/test/anchor_smiles_extended.csv',index=False)
```
