```python
%load_ext autoreload
%autoreload 2
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
import warnings
warnings.filterwarnings("ignore")

import os
import csv

ddir = 'data/'
fname = os.path.join(ddir,'augmented_20to30_10000.csv')

anc_to_augs = {}
with open(fname) as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        anc_to_augs[row[0]] = row[1:]
```

### Augmented SMILES dataframe ... 
Create dataframe of (anchor index), (smiles)

```python
import numpy as np 
import pandas as pd

modeling_ddr = 'model_dataset/'
anc_smileses = sorted([x for x in anc_to_augs.keys()])
anc_smileses.sort(key=lambda x:len(x))
anc_smileses = np.asarray(anc_smileses)

# Save file !!
df_anc_smi = pd.DataFrame(anc_smileses,columns=['smiles'])
df_anc_smi.to_csv(f'{modeling_ddr}anchor_smiles.csv',index=False)
```

```python
from tqdm import tqdm

df_augs = []
for i,anc_smi in tqdm(enumerate(anc_smileses),total=len(anc_smileses)):
    
    aug_smileses = anc_to_augs[anc_smi]
    
    for aug_smi in aug_smileses:
        ser = pd.Series({'anc_idx': i,'smiles':aug_smi}) 
        df_augs.append(ser)
    
df_augs = pd.DataFrame(df_augs)
```

```python
# Save file !!!
df_augs.to_csv(f'{modeling_ddr}augmented_smiles.csv',index=False)
```
