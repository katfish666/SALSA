```python
%load_ext autoreload
%autoreload 2
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
import warnings
warnings.filterwarnings("ignore")

import os
import csv

ddir = 'data/'

fname = os.path.join(ddir,'augmented_20to30_10000.csv')

smiles_to_augs = {}
with open(fname) as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        smiles_to_augs[row[0]] = row[1:]
```

```python
test = next(iter(smiles_to_augs.keys()))
smileses = smiles_to_augs[test]
print(test,smileses)
```

```python
## Just testing that none of the molecules error out ...

from tqdm import tqdm
from property_predictors import surface_predictor

all_props = []

for k,smileses in tqdm(smiles_to_augs.items(),total=len(smiles_to_augs)):
    props = surface_predictor(k)
    all_props.append(props)
    for smiles in smileses:
        props = surface_predictor(smiles)
        all_props.append(props)
```

```python
for p in all_props:
    if sum(p)<1:
        print(p)
        break
```
